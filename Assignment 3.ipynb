```python
import tushare as ts
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
pro = ts.pro_api()
```

```python
def get_profit(ts_code_list: "list")-> 'dictionary':
    '''
    ts_code_list: a list of ts_code

    output:
    a dictionary of dataframes that corresponds to the provided ts_code
    
    备注: netprofit_magin等同于net_profit_ratio
    '''
    l = {}
    for c in ts_code_list:
        l[c] = pro.fina_indicator(ts_code = c)
        l[c] = l[c].query("roe > 15 & netprofit_margin > 30")
    return l
```

```python
# Test case:
df_list = get_profit(['600000.SH', '600003.SH'])
```

```python
# 上证指数
df1 = pro.index_daily(ts_code = '000001.SH')
df1['trade_date'] = pd.to_datetime(df1['trade_date'].astype(str), format = '%Y-%m-%d')
df1
```

```python
# 中信证券
df2 = pro.daily(ts_code = '600030.SH')
df2['trade_date'] = pd.to_datetime(df2['trade_date'].astype(str), format = '%Y-%m-%d')
df2
```

```python
# 华鲁恒升
df3 = pro.daily(ts_code = '600426.SH')
df3['trade_date'] = pd.to_datetime(df3['trade_date'].astype(str), format = '%Y-%m-%d')
df3
```

```python
df1 = df1.loc[:, ['trade_date', 'close']]
df2 = df2.loc[:, ['trade_date', 'close']]
df3 = df3.loc[:, ['trade_date', 'close']]
```

```python
df = df1.merge(df2, on = 'trade_date', how = 'outer').merge(df3, on = 'trade_date', how = 'outer')
```

```python
df.head(5)
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
```

```python
# together
fig, ax = plt.subplots()
ax= sns.lineplot(x='trade_date', y='close_x', data = df)
ax1 = sns.lineplot(x='trade_date', y='close_y', data = df)
ax2 = sns.lineplot(x= 'trade_date', y = 'close', data = df)
ax.set_ylabel('close')
ax.legend(['上证指数', '中信证券', '华鲁恒升'])
```

```python
# separated
fig, axs = plt.subplots(nrows=3, figsize = (15, 15))
sns.lineplot(x='trade_date', y='close_x', data=df, ax=axs[0])
sns.lineplot(x='trade_date', y='close_y', data=df, ax=axs[1], color = 'orange')
sns.lineplot(x='trade_date',y='close', data=df, ax=axs[2], color = 'green')
axs[0].set_title('上证指数日线变化情况')
axs[1].set_title('中信证券日线变化情况')
axs[2].set_title('华鲁恒升日线变化情况')
axs[0].set_ylabel('close')
axs[1].set_ylabel('close')
axs[2].set_ylabel('close')
```
