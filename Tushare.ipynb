# 1. 沪深股票

## 1.1 基础数据

## Pro version:
Tushare的行情等时间序列数据，一般都有两个常用参数：`trade_date`和`ts_code`，分别是交易日期和证券代码。如果你是想提取部分个股的历史数据，用ts_code参数，加上开始和结束日期可以方便提取数据。

## stock_basic

获取基础信息数据，包括股票代码、名称、上市日期、退市日期等

```python
data = pro.query('stock_basic', exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_date')
data
```

## trade_cal

获取各大交易所交易日历数据,默认提取的是上交所

```python
df = pro.trade_cal(exchange='', start_date='20180101', end_date='20181231')
df
```

## namechange

历史名称变更记录

```python
df = pro.namechange(ts_code='600848.SH', fields='ts_code,name,start_date,end_date,change_reason')
df
```

## hs_const

获取沪股通、深股通成分数据

```python
#获取沪股通成分
df1 = pro.hs_const(hs_type='SH') 

#获取深股通成分
df2 = pro.hs_const(hs_type='SZ')
```

```python
df1.head(5)
```

```python
df2.head(5)
```

## stock_company

获取上市公司基础信息，单次提取4500条，可以根据交易所分批提取

```python
df = pro.stock_company(exchange='SZSE', fields='ts_code,chairman,manager,secretary,reg_capital,setup_date,province')
df
```

## new_share

获取新股上市列表数据

```python
df = pro.new_share(start_date='20180901', end_date='20181018')
df
```

## 1.2 行情数据

## daily

A股日线行情。 交易日每天15点～16点之间入库。本接口是未复权行情，停牌期间不提供数据

```python
#多个股票
df = pro.daily(ts_code='000001.SZ,600000.SH', start_date='20180701', end_date='20180718')
df
```

## weekly

获取A股周线行情

```python
df = pro.weekly(ts_code='000001.SZ', start_date='20180101', end_date='20181101', 
                fields='ts_code,trade_date,open,high,low,close,vol,amount')
df
```

## monthly

获取A股月线数据

```python
df = pro.monthly(ts_code='000001.SZ', start_date='20180101', end_date='20181101', 
                 fields='ts_code,trade_date,open,high,low,close,vol,amount')
df
```

## pro_bar

A股复权行情 (日线，周线，月线)

```python
# 日线复权

#取000001的前复权行情
df1 = ts.pro_bar(ts_code='000001.SZ', adj='qfq', start_date='20180101', end_date='20181011')

#取000001的后复权行情
df2 = ts.pro_bar(ts_code='000001.SZ', adj='hfq', start_date='20180101', end_date='20181011')
```

```python
df1
```

```python
df2
```

## adj_factor

获取股票复权因子，可提取单只股票全部历史复权因子，也可以提取单日全部股票的复权因子。

```python
#提取2018年7月18日复权因子
df = pro.adj_factor(ts_code='', trade_date='20180718')
```
